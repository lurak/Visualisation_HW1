```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('GLB.Ts_dSST.csv', skiprows=1) 
df.head(10)
```

```python
df.columns
```

```python
df['mean'] = df.iloc[:, 1:13].mean(axis=1)
df.head()
```

```python
years_list = list()
for item in df['Year']:
    if item % 10 == 0:
        years_list.append(item)
```

```python
def_80_99 = df[(df["Year"] >= 1880) & (df["Year"] <= 1899)]
years_df = def_80_99.mean()['mean']
```

```python
left_point_years = [1944, 2014, 2015, 2017]
left_years = df[df["Year"].isin(left_point_years)]
```

```python
right_point_years = [1904, 1998]
right_years = df[df["Year"].isin(right_point_years)]
```

```python
max_year = df[df["Year"] == 2016]
```

```python
left_text_years = alt.Chart(left_years).mark_text( 
    align='left', 
    dx=10
).encode( 
     text='Year:O') 

right_text_years = alt.Chart(right_years).mark_text( 
    align='right', 
    dx=-10
).encode( 
     text='Year:O') 

text_max_year = alt.Chart(max_year).mark_text( 
    align='left', 
    dx=10, 
    fontWeight='bold' 
).encode( 
    text='Year:O' 
)

main_plot = (alt.Chart(df).mark_point(color='orange', filled=True)+ left_text_years + right_text_years
             + text_max_year ).encode(
    x = alt.X('Year:O', title='', axis=alt.Axis(values=years_list)),
    y = alt.Y('mean:Q', title='')
)


line = alt.Chart(pd.DataFrame({'y': [years_df]})).mark_rule().encode(y=alt.Y('y'))

line_text = alt.Chart(df).mark_text(text=['HOTTER THAN THE', '1880-1899 AVARAGE' ,'COLDER'], dx=370, dy = 88,
                                    color='gray', fontSize=11)

(line_text + line + main_plot).configure_axis(domain=False).configure_axisX(
    labelAngle=0,
    grid=True).properties(
    height = 400,
    width=850,
    title=alt.TitleParams(
        ['Annual Global Surface Temperature,', 'Relative to Late 19th Century Avarage'],
        dx = -275,
        dy = 50,
        fontSize=15
    )
)
```
